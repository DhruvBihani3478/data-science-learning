**Question 19:** Write a Python program to check if a string is a palindrome.

```python
# Your code here
# Solution
string = input("Enter a string: ")
if string == string[::-1]:
    print(f"{string} is a palindrome.")
else:
    print(f"{string} is not a palindrome.")
```
