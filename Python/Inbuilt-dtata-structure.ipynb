### Words Frequency in a Sentence

```python
s='Hello World This is Python World'
s=s.lower()
words_frequency={}
for word in s.split():
    words_frequency[word]=words_frequency.get(word,0)+1
print(words_frequency)
```

### Palindromic Tuple

```python
tpl=(1,2,3,4,3,2,1)
tpl_to_lst=list(tpl)
print(tpl_to_lst[:]==tpl_to_lst[::-1])
```

### Merge Dictionaries with Common Keys

```python
dict3 = {'a': 10, 'b': 20, 'c': 30}
dict4 = {'b': 5, 'c': 15, 'd': 40}
merged = dict3.copy()
for key, value in dict4.items():
    merged[key] = merged.get(key, 0) + value
print(merged)
```
