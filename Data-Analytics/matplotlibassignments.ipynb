### Basic Line Plot

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [2, 5, 3, 8]

plt.plot(x, y)

plt.show()
```

### Create a plot with:Red line,Line width = 4,Square markers,Marker size = 10,Yellow marker face and Black marker edge

```python
x = [1,2,3,4]
y = [1,4,16,25]
plt.plot(x,y,color="red",linewidth=4,marker='s',markersize=10,mfc='yellow',mec='black')
plt.show()
```

### Plot two mathematical functions: y=x²and y=x³
### Add:Title,Axis labels,Legend,Grid

```python
x=[1,2,3,4,5]
y1=[i**2 for i in x]
y2=[i**3 for i in x]
plt.plot(x, y1, label="y = x²",linewidth=2)
plt.plot(x, y2, label="y = x³",linestyle="--")

plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Square and Cube Functions',fontweight='bold')

plt.grid()
plt.legend()
```

### Create a 2×2 subplot figure with:
### Figure title: "Data Visualization"
### Individual titles for each subplot
### Grid enabled on all subplots

```python
import numpy as np

fig,ax=plt.subplots(2,2,figsize=(10,8))
fig.suptitle('Data Visualization')

x = np.linspace(0, 2*np.pi, 100)
ax[0][0].plot(x,np.sin(x))
ax[0][0].set_title('Sine')
ax[0][0].grid()

ax[0][1].plot(x,np.cos(x))
ax[0][1].set_title('Cosine')
ax[0][1].grid()

ax[1][0].plot(x,np.tan(x))
ax[1][0].set_title('Tangent')
ax[1][0].grid()

ax[1][1].plot(x,np.exp(x/5))
ax[1][1].set_title('Exponential')
ax[1][1].grid()


fig.tight_layout()
plt.show()
```

### Other method

```python
x = np.linspace(0, 2*np.pi, 100)
plt.figure(figsize=(10,8))

plt.subplot(2,2,1)
plt.plot(x,np.sin(x))
plt.grid(True)
plt.title('Sine')

plt.subplot(2,2,2)
plt.plot(x,np.cos(x))
plt.grid(True)
plt.title('Cosine')

plt.subplot(2,2,3)
plt.plot(x,np.tan(x))
plt.grid(True)
plt.title('Tangent')

plt.subplot(2,2,4)
plt.plot(x,np.exp(x/5))
plt.grid(True)
plt.title('Exponential')


plt.suptitle('Data Visualization')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [5,10,15,20,25]

plt.figure(figsize=(7,5))

plt.plot(x, y, marker="o")

plt.xlim(0,6)
plt.ylim(0,30)

plt.xticks(
    [1,2,3,4,5],
    ["Mon","Tue","Wed","Thu","Fri"],
    rotation=30
)

plt.yticks([0,5,10,15,20,25,30])

plt.tick_params(
    labelsize=11,
    direction="inout"
)

plt.grid(True)

plt.show()
```
